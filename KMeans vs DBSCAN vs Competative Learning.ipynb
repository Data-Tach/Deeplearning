### Competative Learning

```python
d = make_moons(n_samples=500,shuffle=True, noise=0.1, random_state=0)
data = d[0]
actual_labels = d[1]
```

```python
actual_labels
```

```python
data
```

```python
initial_centroids = np.random.rand(2,2)

c = np.array([1,1])
```

```python
def euclidean_distance(x,y):
    return np.linalg.norm(x-y)
```

```python
data[:,0] = (data[:,0] - np.mean(data[:,0])) / np.std(data[:,0])
data[:,1] = (data[:,1] - np.mean(data[:,1])) / np.std(data[:,1])
```

```python
data
```

```python

epochs = 10

history = []

initial_centroids = np.random.rand(2,2)

for _ in range(epochs):
    c = np.array([1,1])
    labels_ = []
    for x in data:
        d = []
        for centroids in initial_centroids:
    #         print(x,centroids, euclidean_distance(x,centroids))
            d.append(euclidean_distance(x,centroids))
        d = np.array(d)

        min_d = min(d*c)
        l_min = list(d*c).index(min_d)
        c[l_min]+=1
        labels_.append(l_min)
        w_min = initial_centroids[l_min,:]

        delta_w = min_d*(x-w_min)
        w_old = w_min
        w_new = w_old + 0.001*(delta_w)
        initial_centroids[l_min,:] = w_new.copy()
    history.append(c)
    #     print(initial_centroids)
    print(_,c, (sum((actual_labels-labels_)**2))/len(actual_labels))
```
