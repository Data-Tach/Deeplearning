```python
import numpy as np
```

```python
X = np.array([[0,0],
    [0,1],
    [1,0],
    [1,1]])

Y = np.array([0,1,1,0])
```

```python
def sigmoid(x):
    return 1.0/(1.0+np.exp(-x))

def derv_sigmoid(x):
    return x*(1.0-x)
```

```python
w1 = 0.1
w2 = 0.1
w3 = 0.1
w4 = 0.1
w5 = 0.1

lr = 0.1

b1 = 0
b2 = 0

epochs = 100000
loss = []
for _ in range(epochs):
    t = 0
    for i in range(4):
        x1,x2 = X[i]
        y = Y[i]

        # feed_forward_propagation
        h1 = x1*w2 + x2*w3 + b1
        a_h1 = sigmoid(h1)

        y1 = x1*w1 + x2*w4 + a_h1*w5 + b2
        a_y1 = sigmoid(y1)
        
        # error

        err = ((a_y1-y)**2)/2
        t+=((a_y1-y)**2)
        
        # back_propagation
        #---
        del_w1 = (y-a_y1)*(-1)*(derv_sigmoid(a_y1))*(x1)
        del_w4 = (y-a_y1)*(-1)*(derv_sigmoid(a_y1))*(x2)
        del_w5 = (y-a_y1)*(-1)*(derv_sigmoid(a_y1))*(a_h1)
        del_b2 = (y-a_y1)*(-1)*(derv_sigmoid(a_y1))
    
        w1 = w1 - lr*(del_w1)
        w4 = w4 - lr*(del_w4)
        w5 = w5 - lr*(del_w5)
        b2 = b2 - lr*(del_b2)
        #---
        del_w2 = (y-a_y1)*(-1)*(derv_sigmoid(a_y1))*(w5)*(derv_sigmoid(a_h1))*(x1)
        del_w3 = (y-a_y1)*(-1)*(derv_sigmoid(a_y1))*(w5)*(derv_sigmoid(a_h1))*(x2)
        del_b1 = (y-a_y1)*(-1)*(derv_sigmoid(a_y1))*(w5)*(derv_sigmoid(a_h1))
        
        w2 = w2 - lr*(del_w2)
        w3 = w3 - lr*(del_w3)
        b1 = b1 - lr*(del_b1)
    print(t/4)
    loss.append(t/4)
```

```python
def predict(x1,x2):
    return sigmoid(x1*w1 + sigmoid(w2*x1+w3*x2 + b1)*w5 + x2*w4 + b2)
```
